```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook

s1 = signal.lti([(10000*np.pi)**2], [-130.20297681806 + 1j*5231.4754229197, -130.20297681806 - 1j*5231.4754229197], [0.03])
w, mag, phase = signal.bode(s1)

plt.figure()
plt.semilogx(w, mag)  # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase)# Bode phase plot
plt.show()
```
