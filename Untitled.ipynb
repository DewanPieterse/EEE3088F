```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook

s1 = signal.lti([(10000*np.pi)**2], [-130.20297681806 + 1j*5231.4754229197, -130.20297681806 - 1j*5231.4754229197], [0.03])
#Set up the zeros, poles and gain of the LTI system

w, mag, phase = signal.bode(s1)#Populate the cut-off, magnitude and phase arrays

plt.figure()#Populate figure 1
plt.semilogx(w, mag)# Bode magnitude plot
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.figure()#Populate figure 2
plt.semilogx(w, phase)# Bode phase plot
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.show()#Show both plots
```
